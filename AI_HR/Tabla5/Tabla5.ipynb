## Porcentaje de eficiencia de las tecnolog√≠as en RRHH ##

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Datos
data = {
    'Tecnolog√≠a': [
        'Machine\nLearning', 'Procesamiento\nLenguaje\nNatural', 'Big Data', 'Chatbots',
        'An√°lisis\nPredictivo', 'Automatizaci√≥n\nRob√≥tica de\nProcesos', 'Reconocimiento\nFacial'
    ],
    'Uso en RRHH': [
        'Predicci√≥n\nrotaci√≥n', 'An√°lisis\nencuestas', 'Optimizaci√≥n\nselecci√≥n', 'Asistencia\nautomatizada',
        'Identificaci√≥n\ntendencias\ncontrataci√≥n', 'Automatizaci√≥n\ntareas\nadministrativas', 'Control de\nacceso y\nseguridad'
    ],
    'Eficiencia (%)': [76, 80, 88, 75, 82, 90, 78],
}

# Crear DataFrame
df = pd.DataFrame(data)

# Configurar gr√°fico
fig, ax = plt.subplots(figsize=(15, 6))
bars = ax.bar(df['Tecnolog√≠a'], df['Eficiencia (%)'], color='#4682B4', edgecolor='white')  # Azul acero

# Ajustar el eje Y con tramos de 10 en 10
ax.set_yticks(np.arange(0, 101, 10))

# Agregar etiquetas con el uso en RRHH
for bar, uso, eficiencia in zip(bars, df['Uso en RRHH'], df['Eficiencia (%)']):
    ax.text(
        bar.get_x() + bar.get_width()/2, bar.get_height() - 5,
        f"{uso}\n{eficiencia}%", ha='center', va='top', fontsize=10, color='white', fontweight='bold'
    )

# Configuraci√≥n de etiquetas y t√≠tulo
ax.set_xlabel('Tecnolog√≠as', fontweight='bold')
ax.set_ylabel('Eficiencia (%)', fontweight='bold')
ax.set_title('Porcentaje de eficiencia de las tecnolog√≠as en RRHH', fontweight='bold')

# Eliminar la cuadr√≠cula
ax.grid(False)

# Mostrar gr√°fico
plt.show()
```
