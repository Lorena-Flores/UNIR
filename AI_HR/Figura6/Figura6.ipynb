## Comparativa de dimensiones clave en empresas que han implementado la IA frente a aquellas que todav√≠a no lo han hecho ##

```python
import matplotlib.pyplot as plt
import numpy as np

# Dimensiones clave extra√≠das de los cuestionarios
labels = [
    "Automatizaci√≥n",
    "Reducci√≥n de tiempos",
    "Retenci√≥n personalizada",
    "Percepci√≥n positiva",
    "Supervisi√≥n √©tica"
]

# Datos de frecuencias corregidos
empresas_con_ia = [9, 8, 9, 6, 9]
empresas_sin_ia = [1, 2, 3, 5, 3]

# Cierre del pol√≠gono
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
empresas_con_ia += empresas_con_ia[:1]
empresas_sin_ia += empresas_sin_ia[:1]
angles += angles[:1]

# Crear figura y aplicar colores invertidos
fig, ax = plt.subplots(figsize=(7, 7), subplot_kw=dict(polar=True))
azul_oscuro = '#4682B4'  # Empresas con IA
azul_claro  = '#87CEEB'  # Empresas sin IA

# Trazar l√≠neas y √°reas
ax.plot(angles, empresas_con_ia, label="Empresas con IA", color=azul_oscuro, linewidth=2)
ax.fill(angles, empresas_con_ia, color=azul_oscuro, alpha=0.25)

ax.plot(angles, empresas_sin_ia, label="Empresas sin IA", color=azul_claro, linewidth=2)
ax.fill(angles, empresas_sin_ia, color=azul_claro, alpha=0.25)

# Ajustes visuales
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels, fontsize=11)
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
plt.tight_layout()
plt.show()
```
