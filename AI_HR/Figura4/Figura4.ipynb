## Gr√°fico de impacto en el sector de RRHH ##

```python
import matplotlib.pyplot as plt

# Datos
categorias = ["HireVue", "Cornerstone", "Eightfold AI"]
valores = [75, 50, 40]
colores = ["#4682B4", "#5F9EA0", "#87CEEB"]

# Crear gr√°fico de barras
plt.figure(figsize=(8, 6))
bars = plt.bar(range(len(categorias)), valores, color=colores)

# Etiquetas y t√≠tulo
plt.xticks(range(len(categorias)), categorias)  # Nombres en el eje X
plt.ylabel("Porcentaje (%)")
plt.title("Gr√°fico de impacto en el sector de RRHH")

# Agregar etiquetas dentro de las barras
for bar, valor, categoria in zip(bars, valores, ["Reducci√≥n\nen tiempos\nde selecci√≥n", 
                                                 "Incremento\nen retenci√≥n\nde talento", 
                                                 "Mejora\nen la precisi√≥n\nde emparejamiento"]):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height()/2, categoria,
             ha="center", va="center", color="white", fontsize=10, fontweight="bold")

# Eliminar la cuadr√≠cula
plt.grid(False)

# Mostrar gr√°fico
plt.show()
```
