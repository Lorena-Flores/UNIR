## Diagrama de Modelos de Aprendizaje Autom√°tico o Machine Learning ##

```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear un grafo dirigido
G = nx.DiGraph()

# Nodo central
G.add_node("Fundamentos\nde la IA")

# Ramas principales
ramas = ["Aprendizaje\nSupervisado", "Aprendizaje\nNo\nSupervisado", "Aprendizaje\npor\nRefuerzo"]
ejemplos = {
    "Aprendizaje\nSupervisado": ["Clasificaci√≥n\nde curr√≠culums", "Predecir\ndesempe√±o\nfuturo"],
    "Aprendizaje\nNo\nSupervisado": ["Segmentaci√≥n\nde empleados\nseg√∫n perfiles", "Detectar\nanomal√≠as"],
    "Aprendizaje\npor\nRefuerzo": ["Simulaciones\nde din√°micas\nlaborales", "Optimizar\nprocesos\nde selecci√≥n"]
}

for rama in ramas:
    G.add_node(rama)
    G.add_edge("Fundamentos\nde la IA", rama)
    for ejemplo in ejemplos[rama]:
        G.add_node(ejemplo)
        G.add_edge(rama, ejemplo)

# Definir posiciones manualmente para una mejor estructura
pos = {
    "Fundamentos\nde la IA": (0, 3),
    "Aprendizaje\nSupervisado": (-1, 2),
    "Aprendizaje\nNo\nSupervisado": (0, 2),
    "Aprendizaje\npor\nRefuerzo": (1, 2),
    "Clasificaci√≥n\nde curr√≠culums": (-1.2, 1),
    "Predecir\ndesempe√±o\nfuturo": (-0.8, 1),
    "Segmentaci√≥n\nde empleados\nseg√∫n perfiles": (-0.2, 1),
    "Detectar\nanomal√≠as": (0.2, 1),
    "Simulaciones\nde din√°micas\nlaborales": (0.8, 1),
    "Optimizar\nprocesos\nde selecci√≥n": (1.2, 1)
}

# Dibujar el grafo con texto en blanco
plt.figure(figsize=(12, 8))
nx.draw(G, pos, with_labels=True, node_size=9000, node_color="#4682B4", edge_color="gray",
        font_size=12, font_weight="normal", font_color="white", arrows=True, linewidths=1, 
        edgecolors="white", node_shape="s")  # "s" para nodos rectangulares

plt.show()
```
